```python
import urllib.request,sys,time
from bs4 import BeautifulSoup
import requests
import pandas as pd
dt=pd.DataFrame(columns=['Title','Abstract','Date','Authors','keywords'])
```

```python

pagesToGet= 168
i=101
  
for p in range(i,pagesToGet+1):
    print('processing page :', p)
    url = 'https://link.springer.com/search/page/'+str(i)+'?search-within=Journal&facet-journal-id=10489&previous-end-year=2000&facet-start-year=1991&date-facet-mode=between&previous-start-year=1991&facet-end-year=2021&query='
    
    i=i+1
    
    try:
        
        page=requests.get(url)                             
    except Exception as e:                                  
        error_type, error_obj, error_info = sys.exc_info()      
        print ('ERROR FOR LINK:',url)                         
        print (error_type, 'Line:', error_info.tb_lineno)      
        continue                                             
    time.sleep(2)   
    soup=BeautifulSoup(page.text,'html.parser')
    lka =soup.find_all('a',attrs={'class':'title'}) 
     
    for j in lka:
        ur = j['href'].strip()
        ur = 'https://link.springer.com/' + ur
        url1 = ur
        try:
        
            pages=requests.get(ur)                            
    
        except Exception as e:                                   
            error_type, error_obj, error_info = sys.exc_info()      
            print ('ERROR FOR LINK:',ur)                          
            print (error_type, 'Line:', error_info.tb_lineno)     
            continue 
        soups=BeautifulSoup(pages.text,'html.parser')
    

        frame=[]
        upperframe=[]
        filename="NEWS.csv"
        f=open(filename,"w", encoding = 'utf-8')
        headers="Statement,Date\n"
        f.write(headers)
    

        Statement = j.text.strip()

        Abstract = soups.find('div',attrs={'class':'c-article-section__content'})
        Date = soups.find('time').text.strip()
        Authors1 = soups.find_all('li',attrs={'class':'c-article-author-list__item'})
        keywords1 = soups.find_all('li',attrs={'class':'c-article-subject-list__subject'})
        Authors = ""
        keywords = ""
        for k in Authors1:
            Authors += k.find('a').text.strip()
            Authors+=", "
        for x in keywords1:
            keywords+= x.find('span').text.strip()
            keywords+=", "

        if Abstract.find('p') is not None:
            Abstract = Abstract.find('p').text.strip()
            frame.append((Statement,Abstract,Date,keywords,Authors))
            upperframe.extend(frame)
            f.close()
            data=pd.DataFrame(upperframe, columns=['Title','Abstract','Date','keywords','Authors'])
            dt=dt.append(data)
    
```
